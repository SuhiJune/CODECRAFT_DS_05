```python
#VISUALIZING TRAFFIC ACCIDENTS HOTSPOTS

#Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import folium
import webbrowser
from folium.plugins import HeatMap

# Loading dataset
df = pd.read_csv("traffic_accidents.csv")
print("\n TRAFFIC ACCIDENT ANALYSIS ")
print(f"Total Number of Accidents: {len(df)}")

# Visualizing ACCIDENT HOTSPOTS on a heatmap
m = folium.Map(location=[13.05, 80.25], zoom_start=12)
heat_data = df[['Latitude', 'Longitude']].values.tolist()
HeatMap(heat_data).add_to(m)
m.save("accident_hotspots.html")
display(m)
webbrowser.open("accident_hotspots.html")

# Visualizing ACCIDENT DISTRIBUTION BY TIME OF DAY
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Time_of_Day',hue='Time_of_Day', palette="coolwarm")
plt.title('Accident Distribution by Time of Day')
plt.xlabel('Time of Day')
plt.ylabel('Accident Count')
plt.show()

# Visualizing ACCIDENT DISTRIBUTION BY WEATHER CONDITIONS
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Weather',hue='Weather', palette="viridis")
plt.title('Accident Distribution by Weather')
plt.xlabel('Weather Condition')
plt.ylabel('Accident Count')
plt.show()

# Visualizing ACCIDENT DISTRIBUTION BY ROAD CONDITIONS
plt.figure(figsize=(8, 5))
sns.countplot(data=df, x='Road_Condition',hue='Road_Condition', palette='Set2')
plt.title('Accident Distribution by Road Condition')
plt.xlabel('Road Condition')
plt.ylabel('Accident Count')
plt.show()
```
